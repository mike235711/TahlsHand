```python
import subprocess
import chess
import chess.engine
import random
import csv
import os

def start_stockfish_engine(path):
    return chess.engine.SimpleEngine.popen_uci(path)

def get_stockfish_evaluation(engine, board, depth=20):
    info = engine.analyse(board, chess.engine.Limit(depth=depth))
    score = info["score"].white() if board.turn == chess.WHITE else info["score"].black()
    return score.score(mate_score=100000)  # Convert evaluation to a single integer

def no_captures_available(board):
    for move in board.legal_moves:
        if board.is_capture(move):
            return False  # A capture is available
    return True  # No captures available

def play_game(engine, game_data, game_limit=100):
    board = chess.Board()
    moves_made = 0

    while not board.is_game_over() and moves_made < game_limit:
        legal_moves = list(board.legal_moves)
        random_move = random.choice(legal_moves)
        board.push(random_move)

        if no_captures_available(board):
            evaluation = get_stockfish_evaluation(engine, board)
            game_data.append((board.fen(), evaluation))
        
        moves_made += 1

def save_to_csv(game_data, iteration):
    filename = os.path.join("data", f"chess_data_{iteration}.csv")
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["FEN", "Evaluation"])
        writer.writerows(game_data)
    print(f"Data saved to {filename}")

# Specify the path to your Stockfish engine executable
stockfish_path = "/Users/miguel/Documents/Stockfish-master/src/stockfish"

# Ensure data directory exists
os.makedirs("data", exist_ok=True)

# Start the Stockfish engine
engine = start_stockfish_engine(stockfish_path)

# Iterations and number of games per iteration
iterations = 50
num_games = 1000

for iteration in range(1, iterations + 1):
    game_data = []
    print(f"Starting iteration {iteration}")
    for _ in range(num_games):
        print(iteration, _)
        play_game(engine, game_data)
    save_to_csv(game_data, iteration)
    print(f"Iteration {iteration} completed.\n")

# Don't forget to close the engine after you're done
engine.quit()
```
